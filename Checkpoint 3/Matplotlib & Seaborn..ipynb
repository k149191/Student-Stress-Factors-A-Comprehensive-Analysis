Menemukan Data Outliers

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Penanganan Data Outlier

Karena pada kolom noise_level, living_conditions, dan study_load terdapat outliers, sehingga dilakukan penanganan dengan melakukan imputasi.

```python
columns_to_impute = ["noise_level", "living_conditions", "study_load"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menghindari SettingWithCopyWarning dengan menggunakan .loc[df.index, col]
    df.loc[df.index, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
print("\nMenampilkan Persentase Outliers Setelah dilakukan Penanganan")
display(results_df)
```

Kontruksi Data (Construct Data)

Mengkonstruksi data/mengorganisir data dengan menggabungkan
*   'academic_performance', 'study_load', 'teacher_student_relationship', 'future_career_concerns', menjadi academic pressure
*   'noise_level', 'living_conditions', 'safety', 'basic_needs', menjadi environmental pressure
*   'social_support', 'peer_pressure', 'extracurricular_activities', 'bullying' menjadi social pressure
*   'headache', 'blood_pressure', 'sleep_quality', 'breathing_problem', menjadi health risk
*   'anxiety_level', 'self_esteem', 'mental_health_history', 'depression', menjadi mental resilience

```python
df["academic_pressure"] = (df["academic_performance"] + df["study_load"] + df["teacher_student_relationship"] + df["future_career_concerns"]) / 4
df["environmental_pressure"] = (df["noise_level"] + df["living_conditions"] + df["safety"] + df["basic_needs"]) / 4
df["social_pressure"] = (df["social_support"] + df["peer_pressure"] + df["extracurricular_activities"] + df["bullying"]) / 4
df["health_risk"] = (df["sleep_quality"] + df["headache"] + df["blood_pressure"] + df["breathing_problem"]) / 4
df["mental_resilience"] = (df["self_esteem"] + df["anxiety_level"] + df["mental_health_history"] + df["depression"]) / 4

print("\nDataset setelah konstruksi data:")
print(df.head())
```

Reduksi Data (Reduction Value)

Mereduksi data/menghapus data yaitu menghapus "academic_performance", "study_load", "teacher_student_relationship", "future_career_concerns",
                   "noise_level", "living_conditions", "safety", "basic_needs",
                   "social_support", "peer_pressure", "extracurricular_activities", "bullying",
                   "sleep_quality", "headache", "blood_pressure", "breathing_problem",
                   "self_esteem", "anxiety_level", "mental_health_history", "depression"

```python
columns_to_drop = ["academic_performance", "study_load", "teacher_student_relationship", "future_career_concerns",
                   "noise_level", "living_conditions", "safety", "basic_needs",
                   "social_support", "peer_pressure", "extracurricular_activities", "bullying",
                   "sleep_quality", "headache", "blood_pressure", "breathing_problem",
                   "self_esteem", "anxiety_level", "mental_health_history", "depression"]

for column in columns_to_drop:
    if column in df.columns:
        df.drop(columns=[column], inplace=True)

print("\nDataset setelah reduksi data:")
print(df.head())
```

#VISUALISASI

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Histogram

Menampilkan distribusi tingkat stress mahasiswa, menunjukkan bagaimana tingkat stres tersebar di antara mahasiswa. Distribusi tingkat stres cukup merata di antara mahasiswa.

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["stress_level"], bins=10, kde=True, color="maroon")
plt.title("Distribusi Tingkat Stres Mahasiswa")
plt.xlabel("Tingkat Stres")
plt.ylabel("Jumlah Mahasiswa")
plt.show()
```

Pie Chart

Komposisi mahasiswa berdasarkan tingkat stres, menunjukkan persentase mahasiswa di tiap kategori tingkat stres.

```python
plt.figure(figsize=(6, 6))
stress_counts = df["stress_level"].value_counts()
labels = stress_counts.index

plt.pie(stress_counts, labels=labels, autopct="%1.1f%%", colors=["pink", "maroon", "red"])
plt.title("Komposisi Mahasiswa Berdasarkan Tingkat Stres")
plt.show()
```

Scatter Plot

Menampilkan hubungan antara tekanan akademik dan risiko kesehatan.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['academic_pressure'], y=df['health_risk'], alpha=0.6, color="blue")
plt.xlabel("Tekanan Akademik")
plt.ylabel("Risiko Kesehatan")
plt.title("Tekanan Akademik vs Risiko Kesehatan")
plt.show()
```

Heatmap

Menampilkan hubungan antar faktor tekanan mahasiswa, menunjukkan hubungan antara berbagai faktor tekanan. Warna merah tua menunjukkan korelasi positif yang kuat, sedangkan warna lebih terang atau mendekati putih menunjukkan korelasi lebih lemah atau negatif.

```python
plt.figure(figsize=(8, 6))
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True, cmap="OrRd", fmt=".2f", linewidths=0.5)
plt.title("Heatmap Korelasi Antar Faktor Tekanan Mahasiswa")
plt.show()
```

# INSIGHT

Pada Histogram dapat dilihat Distribusi tingkat stres mahasiswa menunjukkan bahwa jumlah mahasiswa dalam kategori stres rendah (0), sedang (1), dan tinggi (2).
Tidak ada dominasi kategori tertentu, yang mengindikasikan bahwa stres di kalangan mahasiswa tersebar merata tanpa kecenderungan signifikan ke salah satu tingkat stres.
Hal ini menunjukkan bahwa sebagian besar mahasiswa mengalami stres dalam berbagai tingkatan, sehingga diperlukan strategi yang dapat menjangkau semua kelompok, baik yang memiliki stres rendah, sedang, maupun tinggi.

Pada Pie Chart dapat dilihat persentase mahasiswa berdasarkan tingkat stresnya. Persentase mahasiswa dalam kategori stres hampir seimbang, dengan sedikit lebih banyak yang berada di kategori 0 (33.9%) dibanding kategori 1 (32.5%) dan kategori 2 (33.5%).
Hal ini menunjukkan bahwa stres mahasiswa tidak didominasi oleh satu tingkat tertentu.

Pada Scatter Plot, dapat dilihat Distribusi data menunjukkan bahwa mayoritas mahasiswa dengan tekanan akademik tinggi juga memiliki risiko kesehatan yang tinggi. Ini bisa mengindikasikan bahwa semakin tinggi beban akademik, semakin besar kemungkinan mahasiswa mengalami masalah kesehatan.
Namun, ada juga mahasiswa dengan tekanan akademik rendah tetapi memiliki risiko kesehatan tinggi. Hal ini menunjukkan bahwa faktor lain seperti tekanan sosial dan lingkungan mungkin turut berkontribusi terhadap kesehatan mahasiswa.
Sebaran data yang tidak sepenuhnya membentuk pola linier menunjukkan bahwa tekanan akademik bukan satu-satunya faktor yang berpengaruh terhadap kesehatan mahasiswa.

Pada Heatmap, dapat dilihat bahwa:
- Tingkat stres memiliki korelasi positif yang kuat dengan tekanan sosial (0.75), yang berarti bahwa semakin tinggi tekanan sosial yang dirasakan mahasiswa, semakin tinggi pula tingkat stres mereka. Hal ini menunjukkan bahwa aspek sosial seperti ekspektasi dari lingkungan, pergaulan, dan interaksi sosial memiliki dampak signifikan terhadap stres mahasiswa.
- Tekanan lingkungan (environmental pressure) memiliki korelasi negatif dengan tingkat stres (-0.69), yang mengindikasikan bahwa lingkungan yang lebih mendukung dan nyaman dapat membantu mengurangi stres mahasiswa. Ini bisa mencakup fasilitas kampus, kondisi tempat tinggal, atau dukungan emosional dari sekitar.
- Tekanan akademik (academic pressure) memiliki korelasi yang sangat lemah dengan tingkat stres (-0.05), yang menunjukkan bahwa tekanan akademik bukan satu-satunya faktor utama yang menyebabkan stres pada mahasiswa. Mahasiswa mungkin telah beradaptasi dengan tekanan akademik atau faktor lain seperti tekanan sosial lebih dominan dalam mempengaruhi stres mereka.
- Ketahanan mental (mental resilience) memiliki korelasi positif dengan tingkat stres (0.43), yang agak kontradiktif karena biasanya ketahanan mental diharapkan membantu mengurangi stres. Hal ini bisa menunjukkan bahwa mahasiswa dengan ketahanan mental tinggi mungkin lebih sadar akan stres mereka, atau mereka mengalami stres dalam jumlah yang lebih tinggi tetapi tetap bisa mengatasinya dengan lebih baik.
