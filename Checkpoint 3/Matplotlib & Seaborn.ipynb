#DATA PREPARATION

Menemukan Nilai Yang Hilang (Missing Value)

```python
print((df.isna().sum() / len(df)) * 100)
```

Menemukan Data Duplikat

```python
df[df.duplicated()]
```

Menemukan Data Outliers

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR)))
print("\nJumlah outliers per kolom:")
print(outliers.sum())
```

Penanganan Data Outlier

```python
for col in ["noise_level", "living_conditions", "study_load"]:
    lower_bound = Q1[col] - 1.5 * IQR[col]
    upper_bound = Q3[col] + 1.5 * IQR[col]

    df[col] = df[col].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))
```

```python
outliers_after = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR)))
print("\nJumlah outliers setelah Winsorizing:")
print(outliers_after.sum())
```

Menememukan Data Yang Tidak Konsisten (Incosistent Value)

```python
print("\nNilai unik per kolom:")
for col in df.columns:
    print(f"{col}: {df[col].unique()}")

print("\nCek nilai negatif:")
print((df < 0).sum())

print("\nRentang nilai per kolom:")
print(df.describe())
```

Penanganan Data Yang Tidak Konsisten (Incosistent Value)

```python
for col in df.columns:
    if col in ["anxiety_level", "depression", "self_esteem", "study_load"]:
        df[col] = df[col].apply(lambda x: abs(x))
```

Kontruksi Data (Construct Data)

```python
df["academic_pressure"] = (df["academic_performance"] + df["study_load"]) / 2
df["social_pressure"] = (df["peer_pressure"] + df["teacher_student_relationship"]) / 2
df["health_risk"] = (df["sleep_quality"] + df["headache"] + df["blood_pressure"] + df["breathing_problem"]) / 4
df["mental_resilience"] = (df["self_esteem"] + df["social_support"]) / 2

print("\nDataset setelah konstruksi data:")
print(df.head())
```

Reduksi Data (Reduction Value)

```python
df.drop(columns=["academic_performance", "study_load", "peer_pressure",
                 "teacher_student_relationship", "sleep_quality", "headache",
                 "blood_pressure", "breathing_problem", "self_esteem", "social_support"], inplace=True)

print("\nDataset setelah reduksi data:")
print(df.head())

df.to_csv("Processed_StressLevelDataset.csv", index=False)
print("\nDataset telah diperbarui dan disimpan sebagai 'Processed_StressLevelDataset.csv'")
```

#VISUALISASI

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

  Bar Chart

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='stress_level', data=df, hue='stress_level', palette="OrRd", legend=False)
plt.title("Distribusi Stress Level Mahasiswa", fontsize=14)
plt.xlabel("Stress Level", fontsize=12)
plt.ylabel("Jumlah Mahasiswa", fontsize=12)
plt.show()
```

Heatmap

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap="OrRd", linewidths=0.5, fmt=".2f")
plt.title("Heatmap Korelasi Antar Variabel", fontsize=14)
plt.show()
```

Boxplot

```python
factors = ["anxiety_level", "depression", "academic_pressure",
           "social_pressure", "health_risk", "mental_resilience"]

plt.figure(figsize=(16, 10))
for i, factor in enumerate(factors, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x="stress_level", y=factor, data=df, hue="stress_level", palette="OrRd", legend=False)
    plt.title(f"Distribusi {factor.replace('_', ' ').title()} Berdasarkan Stress Level", fontsize=12)
    plt.xlabel("Stress Level")
    plt.ylabel(factor.replace('_', ' ').title())

plt.tight_layout()
plt.show()
```

# INSIGHT

Pada Bar Chart dapat dilihat distribusi jumlah mahasiswa dengan stress levelnya. Sehingga dapat diketahui bahwa pada stress level 0 (rendah) terdapat >350 mahasiswa (373 mahasiswa), pada stress level 1 (sedang) terdapat >= 350 mahasiswa (358 mahasiswa), dan pada stress level 2 (tinggi) terdapat >350 mahasiswa (369 mahasiswa).

Pada Heatmap, dapat dilihat korelasi antar variabel. Korelasi Anxiety Level (0.74) dan Depression (0.73) memiliki korelasi paling kuat terhadap tingkat stres mahasiswa. Ini menunjukkan bahwa semakin tinggi tingkat kecemasan dan depresi, semakin tinggi pula tingkat stres mahasiswa. Future Career Concerns (0.74) dan Bullying (0.75) juga berhubungan erat dengan tingkat stres. Kekhawatiran terhadap masa depan dan pengalaman perundungan dapat meningkatkan tekanan psikologis mahasiswa. Mental Resilience (-0.77) memiliki korelasi negatif yang cukup kuat dengan stress level, menunjukkan bahwa mahasiswa dengan ketahanan mental yang baik cenderung memiliki tingkat stres yang lebih rendah.
Safety (-0.71) dan Basic Needs (-0.71) juga memiliki korelasi negatif yang signifikan, yang berarti mahasiswa yang merasa aman dan memiliki kebutuhan dasar yang terpenuhi cenderung mengalami tingkat stres yang lebih rendah. Noise Level (0.69) memiliki hubungan cukup tinggi dengan stress level, yang mengindikasikan bahwa lingkungan yang bising dapat memperburuk kondisi stres mahasiswa.
Living Conditions (-0.71) menunjukkan bahwa kondisi tempat tinggal yang baik berkontribusi pada penurunan tingkat stres.

Pada Boxplot, dapat dilihat hubungan antara stress level dan variabel.
1. Anxiety Level vs. Stress Level
Mahasiswa dengan stress level tinggi (2) memiliki tingkat kecemasan (anxiety) yang jauh lebih tinggi dibandingkan dengan stress level rendah (0).
Semakin tinggi tingkat stres, semakin tinggi pula median anxiety level, dengan banyak outlier yang menunjukkan beberapa mahasiswa memiliki kecemasan ekstrem.
2. Depression vs. Stress Level
Terdapat hubungan positif antara depresi dan stress level.
Mahasiswa dengan stress level tinggi (2) memiliki depresi yang lebih parah, dengan distribusi nilai yang lebih besar dibandingkan level stres rendah (0).
3. Academic Pressure vs. Stress Level
Tekanan akademik cenderung stabil di semua tingkat stress level, tetapi sedikit lebih tinggi pada mahasiswa dengan stress level sedang (1).
Ini menunjukkan bahwa tekanan akademik mungkin bukan faktor utama penyebab stress, dibandingkan faktor lain seperti kecemasan dan depresi.
4. Social Pressure vs. Stress Level
Tekanan sosial memiliki distribusi yang cukup seragam di semua level stres.
Tidak ada perbedaan yang mencolok, yang menunjukkan bahwa tekanan sosial mungkin tidak berkontribusi signifikan terhadap stress level mahasiswa.
5. Health Risk vs. Stress Level
Semakin tinggi stress level, semakin tinggi pula risiko kesehatan mahasiswa.
Mahasiswa dengan stress level tinggi (2) cenderung memiliki health risk lebih besar, kemungkinan karena stres berkepanjangan berdampak pada kesehatan fisik mereka.
6. Mental Resilience vs. Stress Level
Semakin tinggi stress level, semakin rendah mental resilience mahasiswa.
Mahasiswa dengan stress level rendah (0) memiliki ketahanan mental yang jauh lebih baik dibandingkan mereka yang berada di level stres tinggi (2).
Ini mengonfirmasi bahwa mental resilience adalah faktor penting dalam mengurangi tingkat stres, sehingga meningkatkan resiliensi dapat membantu mahasiswa menghadapi stres dengan lebih baik.
